```python
import pandas, keras, numpy, matplotlib.pyplot
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import preprocessing
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
ds = pandas.read_csv("hotel_bookings.csv")
ds.head()
```

```python
dse = ds.dropna()
X = dse.drop(columns=['is_canceled','reservation_status','reservation_status_date'])
X['hotel']=X['hotel'].astype('category')
X['hotel']= X['hotel'].cat.codes
X['arrival_date_month']=X['arrival_date_month'].astype('category')
X['arrival_date_month']= X['arrival_date_month'].cat.codes
X['meal']=X['meal'].astype('category')
X['meal']= X['meal'].cat.codes
X['country']=X['country'].astype('category')
X['country']= X['country'].cat.codes
X['market_segment']=X['market_segment'].astype('category')
X['market_segment']= X['market_segment'].cat.codes
X['distribution_channel']=X['distribution_channel'].astype('category')
X['distribution_channel']= X['distribution_channel'].cat.codes
X['reserved_room_type']=X['reserved_room_type'].astype('category')
X['reserved_room_type']= X['reserved_room_type'].cat.codes
X['assigned_room_type']=X['assigned_room_type'].astype('category')
X['assigned_room_type']= X['assigned_room_type'].cat.codes
X['deposit_type']=X['deposit_type'].astype('category')
X['deposit_type']= X['deposit_type'].cat.codes
X['agent']=X['agent'].astype('category')
X['agent']= X['agent'].cat.codes
X['customer_type']=X['customer_type'].astype('category')
X['customer_type']= X['customer_type'].cat.codes


y = dse['is_canceled']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

classifier = KNeighborsClassifier(n_neighbors=5)
classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
model = LogisticRegression(solver='liblinear', random_state=0)
model.fit(X, y)
print(model.intercept_)
print(model.coef_)
model.predict_proba(X)
print(model.score(X, y))
print(confusion_matrix(y, model.predict(X)))
print(classification_report(y, model.predict(X)))
```
